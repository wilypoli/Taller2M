```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definir la función f(x)
def f(x):
    return x**3 - 3 * x**2 + x - 1

# Definir la derivada de la función f'(x)
def df(x):
    return 3 * x**2 - 6 * x + 1

# Lista de valores iniciales a probar
valores_iniciales = [0, 1, 1.5, 2, -0.5, 2.5]

# Intentar con diferentes valores iniciales
for x0 in valores_iniciales:
    try:
        # Llamada al método de Newton con la función y su derivada
        root = newton(f, x0, fprime=df, tol=1e-6, maxiter=200)
        print(f"Con valor inicial x0 = {x0}, la raíz encontrada es: {root}")
        break  # Salir del bucle si encuentra una raíz
    except RuntimeError as e:
        print(f"Con valor inicial x0 = {x0}, no se logró convergencia.")

# Crear un rango de valores para graficar la función
x = np.linspace(-2, 4, 400)
y = f(x)

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)  # Línea horizontal en y = 0
if 'root' in locals():
    plt.axvline(root, color="red", linestyle="--", label=f"Raíz en x = {root:.5f}")  # Marcar la raíz en la gráfica
    plt.scatter(root, f(root), color="red", s=50, label="Punto de raíz")  # Agregar el punto rojo en la raíz
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Gráfica de la ecuación $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
